Train, Test & Split

```python
from sklearn.model_selection import train_test_split
```

```python
X = data.drop("price",axis=1)
y = data["price"]
```

```python
X.head()
```

```python
y.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.33, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lrg = LinearRegression()
```

```python
lrg.fit(X_train,y_train)
```

```python
lrg.score(X_test,y_test)
```

lrg.score(X_test,y_test) : 0.6966159937240239 

```python
from sklearn import ensemble
from sklearn.metrics import mean_squared_error
```

```python
params = {
    "n_estimators": 400,
    "max_depth": 5,
    "min_samples_split": 2,
    "learning_rate": 0.1,
    "loss": "squared_error",
    "verbose": 3
}
```

```python
model = ensemble.GradientBoostingRegressor(**params)
```

```python
model.fit(X_train, y_train)
```

```python
model.score(X_test,y_test)
```

#Gradient boosting regression algorithm over predictive model yield an R2 of 87.2%. As compaired to accuracy of 69.66% predictive accuracy by Linear regression model.

Mean squared error and Root mean squared error for gradient boosting regression 

```python
mse = mean_squared_error(y_test, model.predict(X_test))
print(f"The mean squared error (MSE) on test set: {mse:.4f}")
```

```python
# Root_mean_squared_error
rmse = np.sqrt(mse)
```

```python
print(f"The root mean squared error (MSE) on test set: {rmse:.4f}")
```

```python
data["price"].describe()
```

Plotting the deviance between test_score and predictive scores

```python
test_score = np.ones((params["n_estimators"]), dtype=np.float64)
```

```python
test_score
```

```python
for i, y_pred in enumerate(model.staged_predict(X_test)):
    test_score[i] = mean_squared_error(y_test, y_pred)
```

```python
test_score
```

```python
model.train_score_
```

```python
plt.figure(figsize=(6, 6))
plt.subplot()
plt.title("Deviance")
plt.plot(
    range(1,401),
    model.train_score_,
    "b-",
    label="Training Set Deviance",
)
plt.plot(
    range(1,401), test_score, "r-", label="Test Set Deviance"
)
plt.legend(loc="upper right")
plt.xlabel("Boosting Iterations")
plt.ylabel("Deviance")
plt.tight_layout()
plt.show()
```

Feature Importance

```python
feature_importance = model.feature_importances_
sorted_idx = np.argsort(feature_importance)
pos = np.arange(sorted_idx.shape[0]) + 0.5
fig = plt.figure(figsize=(12, 6))
plt.barh(pos, feature_importance[sorted_idx], align="center")
plt.yticks(pos, np.array(data.columns)[sorted_idx])
plt.title("Feature Importance")
```
